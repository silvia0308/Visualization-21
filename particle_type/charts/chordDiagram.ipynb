```python
import pandas as pd
from chord import Chord
import numpy as np
```

```python
names = ["other baryons","star-forming","star","wind","dark matter","AGN"]
```

```python
with open('Distribute/Trans.txt') as f:
    lines = [line.rstrip() for line in f]
```

```python
matrix=[]
for i in range(6):
    row = lines[i].split()
    for j in range(6):
        row[j] = np.log( float(row[j]) + 1 )
    matrix.append(row)
matrix
```
